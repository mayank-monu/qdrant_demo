```python
from sentence_transformers import SentenceTransformer
import numpy as np
import json
import pandas as pd
from tqdm.notebook import tqdm
```

```python
model = SentenceTransformer('distilbert-base-nli-stsb-mean-tokens', device="cuda")
```

```python
df = pd.read_json('./anthem_json/depr.json',lines=True)
```

```python
vectors = []
batch_size = 64
batch = []
for row in tqdm(df.itertuples()):
  question = row.prompt
  batch.append(question)
  if len(batch) >= batch_size:
    vectors.append(model.encode(batch))  # Text -> vector encoding happens here
    batch = []

if len(batch) > 0:
  vectors.append(model.encode(batch))
  batch = []

vectors = np.concatenate(vectors)
```

```python
np.save('./anthem_json/vectors.npy', vectors, allow_pickle=False)
```
